```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
```

```
X,y= make_moons(n_samples=200, noise=0.10)
```

```
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```
kmeans=KMeans(n_clusters=2)
kmeans.fit(X,y)
```

```
center=kmeans.cluster_centers_
```

```
pred=kmeans.labels_
```

```
plt.scatter(X[:,0],X[:,1],c=pred)
plt.scatter(center[:,0],center[:,1],c='black',marker='*')
plt.show()
```

**DBSCAN**
* density spatial clustering of Application with noise

* based upon idea that a cluster is a high density area surrounded by low density region
* starts by exploring a small area, if density is 'good enough' it is considered as part of cluster and explore neighbors and so on so it group region based upon density
* i.e. it starts with one point and surrounds all other nearby points based on minimum distance

```
from sklearn.cluster import DBSCAN
```

```
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```
db= DBSCAN?
```

```
db=DBSCAN(eps=0.2, min_samples=5)
```

```
db.fit(X)
```

```
ypred=db.fit_predict(X)
print(ypred)
```

```
plt.scatter(X[:,0],X[:,1],c=ypred)

plt.show()
```
