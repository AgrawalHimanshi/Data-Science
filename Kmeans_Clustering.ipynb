```
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```
X,y=make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
X.shape
```

```
plt.figure()
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```
k=5
All_cluster={}
color=["red","green","blue","orange","yellow"]

for kx in range(k):
  center=10*(2*np.random.random((X.shape[1],))-1)
  points=[]
  cluster={
   'center':center,
   'points':points,
   'color':color[kx],   
  }
  All_cluster[kx]=cluster

  
```

```
# center=10*(2*np.random.random((X.shape[1],))-1)
# print(center)
All_cluster
```

```
def distance(v1,v2):
  return np.sqrt(np.sum((v1-v2)**2))
```

```
# E-step
def assignPointsToCluster(All_cluster):
  for i in range(X.shape[0]):
    curr_x=X[i]
    distlist=[]
    for kx in range(k):
      dist=distance(curr_x, All_cluster[kx]['center'])
      distlist.append(dist)
    req_cluster=np.argmin(distlist)
    All_cluster[req_cluster]['points'].append(curr_x)

def plotPoints(All_cluster):
  for kx in range(k):
    # print(len(All_cluster[kx]['points']))
    pts=np.array(All_cluster[kx]['points'])
    print(pts.shape)
    # plot points
    try:
      plt.scatter(pts[:,0],pts[:,1],color=All_cluster[kx]['color'])
    except:
      pass
    # plot cluster center
    uk=All_cluster[kx]['center']
    plt.scatter(uk[0],uk[1], color='black',marker='*')

def UpdateCluser(All_cluster):
  for kx in range(k):
    pts=np.array(All_cluster[kx]['points'])
    if pts.shape[0]>0:
      new_uk=pts.mean(axis=0)
      All_cluster[kx]['center']=new_uk
      All_cluster[kx]['points']=[]
      
```

```
assignPointsToCluster(All_cluster)
plotPoints(All_cluster)
UpdateCluser(All_cluster)
```

```
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```
X,y=make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
X.shape
```

```
kmeans=KMeans(n_clusters=5)
```

```
kmeans.fit(X,y)
```

```
center=kmeans.cluster_centers_
```

```
plt.scatter(X[:,0],X[:,1])
plt.scatter(center[:,0],center[:,1],color='black',marker='*')
plt.show()
```

```
pred=kmeans.labels_
```

```
plt.scatter(X[:,0],X[:,1],c=pred)
plt.scatter(center[:,0],center[:,1],color='black',marker='*')
plt.show()
```
